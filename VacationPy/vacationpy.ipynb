```python
# Dependencies and Setup
import gmaps
import json
import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd
import requests

# Import API key
from config import gkey
```

```python
# import weather CSV from part one into a df
dataFilepath = "../WeatherPy/weatherDataDF.csv"
weatherDataDF = pd.read_csv(dataFilepath)
weatherDataDF
```

```python
# Humidity Heatmap

# Configure gmaps.
gmaps.configure(api_key=gkey)

# Use the Lat and Long as locations and Humidity as the weight.
locations = weatherDataDF[["Latitude", "Longitude"]].astype(float)
weight = weatherDataDF["Relative Humidity"].astype(int)

# fig = gmaps.figure()
fig = gmaps.figure(zoom_level = 1.6, center = (20, 5))

# Add Heatmap layer to map.

heat_layer = gmaps.heatmap_layer(locations, weights=weight, 
                                 dissipating=False, max_intensity = 100,
                                 point_radius = 3)

fig.add_layer(heat_layer)

fig
```

```python
# Narrow down the DataFrame to find your ideal weather condition. For example:
# A max temperature lower than 80 degrees but higher than 70.
# Wind speed less than 10 mph.
# Zero cloudiness.
# Drop any rows that don't contain all three conditions.
# You want to be sure the weather is ideal.
# Note: Feel free to adjust your specifications but be sure to
# limit the number of rows returned by your API requests to a reasonable number; e.g., 10.
# Use Google Places API to find the first hotel for each city
# located within 5000 meters of your coordinates.
# Plot the hotels on top of the humidity heatmap with each pin containing
# the Hotel Name, City, and Country.

# Create new DataFrame fitting weather criteria
# Narrow down the cities to fit weather conditions.

minTemp = 15
maxTemp = 32
minCloudy = 90
maxCloudy = 100

hotels_df = weatherDataDF.loc[(weatherDataDF["Temperature"] < maxTemp) & (weatherDataDF["Temperature"] > minTemp) & (weatherDataDF["Cloudiness"] > minCloudy) & (weatherDataDF["Cloudiness"] < maxCloudy), :]   
print(len(hotels_df))
hotels_df
```

```python
# Drop any rows with null values.
# There aren't any.

# Hotel Map

# Store into variable named hotel_df.

# Add a "Hotel Name" column to the DataFrame.
hotels_df["Hotel Name"] = ""
hotels_df["City"] = ""
hotels_df["Country"] = ""
hotels_df
```

```python
placeCount = 0
print(f"At time of measurement, these geocoordinates had temps between {minTemp}\N{DEGREE SIGN}F and {maxTemp}\N{DEGREE SIGN}F and skies between {minCloudy}% and {maxCloudy}% cloudy:")
for place in hotels_df["Temperature"]:
    try:
        # read Latitude out of row placeCount in hotels_df
        searchLat = hotels_df.iloc[placeCount, 1]

        # read Longitude out of row placeCount in hotels_df
        searchLong = hotels_df.iloc[placeCount, 2]

        # PUfD
        # print(f"{searchLat}, {searchLong}")

        # Hit the Google Places API for each city's coordinates.

        # geocoordinates
        target_coordinates = f"{searchLat}, {searchLong}"
        target_search = "hotel"
        target_radius = 5000
        target_type = "hotel"

        # PUfD
        print(f"{searchLat} lat., {searchLong} long")

        # set up a parameters dictionary
        params = {
            "location": target_coordinates,
            "keyword": target_search,
            "radius": target_radius,
            "type": target_type,
            "key": gkey
        }

        # base url
        base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

        # run a request using our params dictionary
        response = requests.get(base_url, params=params)
        # print(response)
        # convert response to json
        places_data = response.json()

        # PUfD the json (pretty printed)
        # print(json.dumps(places_data, indent=4, sort_keys=True))

        hotelName = places_data["results"][0]["name"]
        # hotelName
        # print(hotelName)
        print(f"hotel: {hotelName.title()}")
        # set "Hotel Name" in hotels_df to hotelName
        hotels_df.iloc[placeCount, hotels_df.columns.get_loc("Hotel Name")] = hotelName


        place = places_data["results"][0]["plus_code"]["compound_code"]
        # place
        # print(f"{place}\n")

        chunks = place.split(",")
        # print(chunks)

        smallerChunks = chunks[(0)]
        smallerChunks = smallerChunks.split(" ", 1)
        city = smallerChunks[(1)]

        print(f"city: {city}")
        hotels_df.iloc[placeCount, hotels_df.columns.get_loc("City")] = city

        country = chunks[(1)]
        country = country.strip()
        print(f"country: {country}\n")
        hotels_df.iloc[placeCount, hotels_df.columns.get_loc("Country")] = country

        placeCount += 1
    except:
        print(f"Sorry, Google couldn't find a hotel within 3.1 miles of {searchLat} lat., {searchLong} long.\n")
        placeCount += 1
```

```python
# This recreates the map with a heat layer for humidity:
fig = gmaps.figure(zoom_level = 1.8, center = (25, 0))
fig.add_layer(heat_layer)

# after reading the nuclear power plants example at https://jupyter-gmaps.readthedocs.io/en/latest/api.html
# this is a test making list of dictionaries by hand and feeding it into the marker layer--works fine:
# hotels = [
#     {"name": "Zevlův mlýn", "location": (48.7896, 14.7784), "city": "Nové Hrady", "country": "Czechia"},
#     {"name": "гостиница Гаралы", 'location': (44.0834, 41.9763), "city": "Ust-Dzheguta", "country": "Karachay-Cherkessia"},
# ]

# Making a list of dictionaries structured exactly as above, but automated:
hotels = []
placeCount = 0
for location in locationsWithHotels_df:
    hotel = locationsWithHotels_df.iloc[placeCount, locationsWithHotels_df.columns.get_loc("Hotel Name")]
    latitude = locationsWithHotels_df.iloc[placeCount, locationsWithHotels_df.columns.get_loc("Latitude")]
    longitude = locationsWithHotels_df.iloc[placeCount, locationsWithHotels_df.columns.get_loc("Longitude")]
    city = locationsWithHotels_df.iloc[placeCount, locationsWithHotels_df.columns.get_loc("City")]
    country = locationsWithHotels_df.iloc[placeCount, locationsWithHotels_df.columns.get_loc("Country")]
#    hotels.append({f"'name': {hotel}, 'location': '('{latitude}, {longitude}')', 'city': {city}, 'country': {country}"})
#    hotels.append(f"'name': {hotel}, 'location': '('{latitude}, {longitude}')', 'city': {city}, 'country': {country}")
    hotels.append({'name': hotel, 'location': (latitude, longitude), 'city': city, 'country': country})
    placeCount += 1
print(hotels)

hotel_locations = locations


info_box_template = """
<dl>
<dt>Hotel Name</dt><dd>{name}</dd>
<dt>City</dt><dd>{city}</dd>
<dt>Country</dt><dd>{country}</dd>
</dl>
"""

hotel_info = [info_box_template.format(**hotel) for hotel in hotels]

marker_layer = gmaps.marker_layer(hotel_locations, info_box_content=hotel_info)
fig.add_layer(marker_layer)
fig
```
